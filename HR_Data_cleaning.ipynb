```python
import pandas as pd
```

```python
df = pd.read_csv(r"C:\Users\divya\Downloads\WA_Fn-UseC_-HR-Employee-Attrition.csv")
df.head()
```

```python
df.isnull().sum()
```

```python
df.duplicated().sum()
df.drop_duplicates(inplace=True)
```

```python
df['AttritionFlag'] = df['Attrition'].apply(lambda x: 1 if x=='Yes' else 0)
```

```python
df['Tenure'] = df['YearsAtCompany']
```

```python
df['AgeGroup'] = pd.cut(df['Age'], 
                        bins=[18,25,35,45,60],
                        labels=['18-25','26-35','36-45','46-60'])
```

```python
df.drop(['EmployeeCount','EmployeeNumber','StandardHours','Over18'], axis=1, inplace=True)
```

```python
df.to_csv("cleaned_hr_data.csv", index=False)
```
